```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({
    "figure.figsize": [6,4], 
})

# Definición de las funciones de activación
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# Rango de valores para el eje X
x = np.linspace(-10, 10, 100)

# Mostrar y guardar función Sigmoide
plt.figure()
plt.plot(x, sigmoid(x), color='blue')
plt.text(-9.5, 0.5, r'$f(x) = \frac{1}{1 + e^{-x}}$', fontsize=20, verticalalignment='center', horizontalalignment='left')
plt.xlim(-10, 10)
plt.ylim(-0.1, 1.1)
plt.show()

# Mostrar y guardar función Tanh
plt.figure()
plt.plot(x, tanh(x), color='green')
plt.text(-9.5, 0, r'$f(x) = \tanh(x)$', fontsize=20, verticalalignment='center', horizontalalignment='left')
plt.xlim(-10, 10)
plt.ylim(-1.1, 1.1)
plt.show()

# Mostrar y guardar función ReLU
plt.figure()
plt.plot(x, relu(x), color='red')
plt.text(-9.5, 1, r'$f(x) = \max(0, x)$', fontsize=20, verticalalignment='center', horizontalalignment='left')
plt.xlim(-10, 10)
plt.ylim(-1, 10.1)
plt.show()

# Mostrar y guardar función Softmax (solo clase 1)
x_softmax = np.array([x, np.ones_like(x), -x])  # Ejemplo con tres clases
y_softmax = softmax(x_softmax)
plt.figure()
plt.plot(x, y_softmax[0], color='orange')
plt.text(-9.5, 0.5, r'$f(x_i) = \frac{e^{x_i}}{\sum_{j} e^{x_j}}$', fontsize=20, verticalalignment='center', horizontalalignment='left')
plt.xlim(-10, 10)
plt.ylim(-0.1, 1.1)
plt.show()


```
