```python
import numpy as np
import matplotlib.pyplot as plt

# Nmero de curvas a generar
n = 4  # Puedes cambiar el valor de n

# Rango de frecuencias omega (ajustado)
w_values = np.linspace(2.09, 2.14, 2500)  # 2500 puntos entre 2.09 y 2.14 GHz

# Parmetros constantes
Omega = 2.116451  # Valor constante de Omega

# Generacin de los parmetros aleatorios dentro de los rangos especificados
k = np.random.uniform(1e-4, 1e-3, n)
kc = np.random.uniform(1e-5, 1e-4, n)
gamma = np.random.uniform(5e-3, 8e-3, n)
g = np.random.uniform(1e-2, 1.5e-2, n)
DeltaE = np.random.uniform(2.09, 2.13, n)

# Funcin para calcular S21
def S21(omega, k, kc, g, DeltaE, gamma, Omega):
    numerator = kc
    denominator = 1j * (Omega - omega) + k + (g**2 / (1j * (DeltaE - omega) + gamma))
    return 1 - numerator / denominator

# Generacin de las curvas y graficacin
plt.figure(figsize=(6, 4))
for i in range(n):
    S21_curve = S21(w_values, k[i], kc[i], g[i], DeltaE[i], gamma[i], Omega)
    plt.plot(w_values, np.abs(S21_curve), label=f'Curva {i+1}')

plt.xlim(2.09,2.14)

plt.show()

```

```python
def t_w (w, k, Omega, DeltaE, gamma, g):
    up = -k
    down = (Omega-w) - 1j * k - 1j * g**2 / (1j * (DeltaE - w) + gamma)
    t = up / down
    return np.abs(t)

# Nmero de curvas a generar
n = 3 # Puedes cambiar el valor de n

# Rango de frecuencias omega (ajustado)
w_values = np.linspace(0.85, 1.15, 300)  # 2500 puntos entre 2.09 y 2.14 GHz

# Parmetros constantes
Omega = 1  # Valor constante de Omega

# Generacin de los parmetros aleatorios dentro de los rangos especificados
k = np.random.uniform(1e-4, 1e-2, n)
gamma = np.random.uniform(1e-4, 1e-2, n)
g = np.random.uniform(1e-4, 1e-2, n)
DeltaE = np.random.uniform(0.98, 1.02, n)

# Generacin de las curvas y graficacin
plt.figure(figsize=(6, 4))
for i in range(n):
    tw_curve = t_w(w_values, k[i], Omega, DeltaE[i], gamma[i], g[i])
    plt.plot(w_values, np.abs(tw_curve), label=f'Curva {i+1}')

plt.xlim(0.85,1.15)
plt.xlabel("$\omega$")
plt.ylabel("$t (\omega)$")

plt.show()
```
